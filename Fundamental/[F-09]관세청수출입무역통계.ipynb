**이상치(outlier)**란 '대부분 값의 범위에서 벗어나 극단적으로 크거나 작은 값'을 의미한다.<br>
#### 이상치를 찾는 방법(anomaly detection)
- 평균을 빼주고 표준편차로 나눠 **z-score$({\frac {X-\mu }{\sigma }})$**를 계산한다.<br> z-score가 특정 기준을 넘어서는 데이터에 대해 이상치라고 판단한다. 기준을 작게 할수록 이상치라고 판단하는 데이터가 많아진다.
- 사분위범위수 **IQR**(Interquartile range)

#### 이상치를 처리하는 방법
- 이상치를 원래 데이터에서 **삭제**하고, 이상치끼리 따로 분석한다.

- 이상치를 다른 값으로 **대체**한다. **데이터가 적으면** 이상치를 삭제하기보다 다른 값으로 대체하는 것이 나을 수 있다. <br>예를 들어 최댓값, 최솟값을 설정해 데이터의 범위를 제한할 수 있습니다.

- 다른 데이터를 활용하여 예측 모델을 만들어 **예측값으로 대체**한다.
- **binning(구간화)을 통해 수치형 데이터를 범주형으로** 바꾼다.

#### z-score 사용

```python
# 표준화한 값이 z보다 큰 데이터의 인덱스 리턴
def outlier(df, col, z):
    return df[abs(df[col] - np.mean(df[col]))/np.std(df[col])>z].index

trade.loc[outlier(trade, '무역수지', 1.5)]
```

```python
# 기준을 높일수록 이상치가 적어진다.
trade.loc[outlier(trade, '무역수지', 2)]
```

```python
# 이상치가 아닌 데이터의 인덱스 리턴
def not_outlier(df, col, z):
    return df[abs(df[col] - np.mean(df[col]))/np.std(df[col]) <= z].index

trade.loc[not_outlier(trade, '무역수지', 1.5)]
```

### z-score의 단점
1. Robust하지 못하다. (= **이상치 값에 영향을 받는다.**) <br>평균과 표준편차 자체가 이상치의 존재에 크게 영향을 받기 때문이다.
2. **작은 데이터셋의 경우** z-score 방법으로 이상치를 알아내기 어렵다.<br>특히 item이 12개 이하인 데이터셋에서는 불가능하다.

[참고](http://colingorrie.github.io/outlier-detection.html)

#### IQR 사용
IQR은 제3사분위수에서 제1사분위수를 뺀 값(Q3 - Q1)이다.<br>
이는 데이터의 50%범위이다.<br>
따라서 Q1-1.5\*IQR보다 왼쪽에 있거나 Q3+1.5\*IQR보다 오른쪽에 있으면 이상치로 판단한다.
![이미지](https://drive.google.com/uc?export=view&id=19XME4Ny-voY_65IWpI5KYSA3QPPvAmya)

```python
# 예제)
np.random.seed(2020)
data = np.random.randn(100)  # 평균 0, 표준편차 1의 분포에서 100개의 숫자를 샘플링한 데이터 생성
data = np.concatenate((data, np.array([8, 10, -3, -5])))      # [8, 10, -3, -5])를 데이터 뒤에 추가함
data
```

```python
# 박스 플롯에서 크게 벗어난 점들이 이상치
fig, ax = plt.subplots()
ax.boxplot(data)
plt.show()
```

> 백분위수(Percentile)는 오름차순 정렬했을 때 백분율로 나타낸 특정 위치 값이다. 따라서 사분위수는 25, 50, 75를 기준 점으로 나눠져 1분위부터 4분위까지 존재하게 된다. 백분위수 q 값은 0부터 100 값을 사용한다.

```python
Q3, Q1 = np.percentile(data, [75 ,25])
IQR = Q3 - Q1
IQR
```

```python
data[(Q1-1.5*IQR > data)|(Q3+1.5*IQR < data)]
```

```python
# IQR로 이상치 판단
def outlier2(df, col):
    q1 = df[col].quantile(0.25)       # quantile을 사용하여 사분위수를 쉽게 구할 수 있다.
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    return df[(df[col] < q1-1.5*iqr)|(df[col] > q3+1.5*iqr)]

outlier2(trade, '무역수지')
```

### 정규화(Normalization) 하는 방법
1. **표준화(Standardization):** 데이터의 평균은 0, 분산은 1이 되도록 변환 $({\frac {X-\mu }{\sigma }})$
2. **Min-Max Scaling:** 데이터의 최솟값이 0, 최댓값이 1이 되도록 변환 $({\frac {X-X_{min}}{X_{max}-X_{min}}})$

[[참고: normalization/standardization/regularization]](https://realblack0.github.io/2020/03/29/normalization-standardization-regularization.html)

```python
# 예제)
# 정규분포를 따라 랜덤하게 데이터 x를 생성합니다. 
np.random.seed(2020)
x = pd.DataFrame({'A': np.random.randn(100)*4+4,
                 'B': np.random.randn(100)-1})
x
```

```python
# 데이터 x를 Standardization 기법으로 정규화합니다. 
x_standardization = (x - x.mean())/x.std()
x_standardization
```

```python
# 데이터 x를 min-max scaling 기법으로 정규화합니다. 
x_min_max = (x-x.min())/(x.max()-x.min())
x_min_max
```

```python
# 데이터를 Standardization 기법으로 정규화를 했을 때 분포가 어떻게 바뀌는지 보여준다.
# 즉, 각 컬럼의 평균은 0으로, 분산은 1로 데이터를 바꿔준다.
fig, axs = plt.subplots(1,2, figsize=(12, 4),
                        gridspec_kw={'width_ratios': [2, 1]})

axs[0].scatter(x['A'], x['B'])
axs[0].set_xlim(-5, 15)
axs[0].set_ylim(-5, 5)
axs[0].axvline(c='grey', lw=1)
axs[0].axhline(c='grey', lw=1)
axs[0].set_title('Original Data')

axs[1].scatter(x_standardization['A'], x_standardization['B'])
axs[1].set_xlim(-5, 5)
axs[1].set_ylim(-5, 5)
axs[1].axvline(c='grey', lw=1)
axs[1].axhline(c='grey', lw=1)
axs[1].set_title('Data after standardization')

plt.show()
```

```python
# 동일한 데이터를 min-max scaling 기법으로 정규화를 했을 때 분포가 어떻게 바뀌는지 보여준다.
# 즉, 각 컬럼의 최솟값은 0, 최댓값은 1로 바꿔준다.
fig, axs = plt.subplots(1,2, figsize=(12, 4),
                        gridspec_kw={'width_ratios': [2, 1]})

axs[0].scatter(x['A'], x['B'])
axs[0].set_xlim(-5, 15)
axs[0].set_ylim(-5, 5)
axs[0].axvline(c='grey', lw=1)
axs[0].axhline(c='grey', lw=1)
axs[0].set_title('Original Data')

axs[1].scatter(x_min_max['A'], x_min_max['B'])
axs[1].set_xlim(-5, 5)
axs[1].set_ylim(-5, 5)
axs[1].axvline(c='grey', lw=1)
axs[1].axhline(c='grey', lw=1)
axs[1].set_title('Data after min-max scaling')

plt.show()
```

```python
# trade 데이터를 Standardization 기법으로 정규화합니다. 
cols = ['수출건수', '수출금액', '수입건수', '수입금액', '무역수지']        # 수치형 데이터를 가진 칼럼들
trade_Standardization= (trade[cols]-trade[cols].mean())/trade[cols].std()
trade_Standardization.head()
```

`cut`: 구간을 일정하게 나눠준다.<br>
**pd.cut(데이터, bins=구간)** 또는 **pd.cut(데이터, bins=구간 수)**
<br><br>
`qcut`: 구간을 일정하게 나누는 것이 아니라 데이터의 분포를 **비슷한 크기의 그룹으로** 나눠줍니다.

```python
bins = [0, 2000, 4000, 6000, 8000, 10000]

# 데이터를 구간별로 나눈다.
ctg = pd.cut(salary, bins=bins)
ctg
```

```python
# 데이터와 카테고리 확인
print('salary[0]:', salary[0])
print('salary[0]가 속한 카테고리:', ctg[0])
```

```python
# 각 구간에 속한 데이터 개수 확인
ctg.value_counts().sort_index()
```

```python
# 데이터를 6개 구간으로 나눈다.
ctg = pd.cut(salary, bins=6)
ctg
```

```python
# 각 구간에 속한 데이터 개수 확인
ctg.value_counts().sort_index()
```

```python
# qcut을 사용하여 5개 구간으로 나눈다.

ctg = pd.qcut(salary, q=5)
ctg
```

```python
print(ctg.value_counts().sort_index())
```
